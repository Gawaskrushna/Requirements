```python
import math
from collections import Counter
# Training data: (IMDB rating, Duration, Genre)
movies = [
    (8.0, 160, "Action"),
    (6.2, 170, "Action"),
    (7.2, 168, "Comedy"),
    (8.2, 155, "Comedy")
]
# Movie to predict
barbie = (7.4, 114)
# Function to calculate Euclidean distance
def euclidean_distance(movie1, movie2):
    return math.sqrt((movie1[0] - movie2[0]) ** 2 + (movie1[1] - movie2[1]) ** 2)
# Compute distances
distances = []
for m in movies:
    dist = euclidean_distance(barbie, (m[0], m[1]))
    distances.append((dist, m[2]))
# Sort by distance
distances.sort(key=lambda x: x[0])
# Choose k nearest neighbors
k = 2
nearest_genres = [genre for _, genre in distances[:k]]
# Perform majority vote
genre_prediction = Counter(nearest_genres).most_common(1)[0][0]
print("Predicted Genre for Barbie:", genre_prediction)
```

```python
import math

# Function to calculate entropy
def entropy(data, target_attr):
    values = [record[target_attr] for record in data]
    unique_vals = set(values)
    ent = 0.0
    for val in unique_vals:
        prob = values.count(val) / len(values)
        ent -= prob * math.log2(prob)
    return ent

# Function to calculate information gain
def info_gain(data, attr, target_attr):
    total_entropy = entropy(data, target_attr)
    values = set([record[attr] for record in data])
    weighted_entropy = 0.0
    for val in values:
        subset = [record for record in data if record[attr] == val]
        weighted_entropy += (len(subset) / len(data)) * entropy(subset, target_attr)
    return total_entropy - weighted_entropy

# ID3 algorithm
def id3(data, attributes, target_attr):
    values = [record[target_attr] for record in data]

    # If all records have same class -> return leaf
    if values.count(values[0]) == len(values):
        return values[0]

    # If no attributes left -> return majority class
    if not attributes:
        return max(set(values), key=values.count)

    # Choose best attribute (highest info gain)
    best_attr = max(attributes, key=lambda attr: info_gain(data, attr, target_attr))

    tree = {best_attr: {}}
    for val in set([record[best_attr] for record in data]):
        subset = [record for record in data if record[best_attr] == val]
        if not subset:
            tree[best_attr][val] = max(set(values), key=values.count)
        else:
            new_attrs = [a for a in attributes if a != best_attr]
            subtree = id3(subset, new_attrs, target_attr)
            tree[best_attr][val] = subtree
    return tree

# Function to visualize tree (text-based)
def print_tree(tree, indent=""):
    if isinstance(tree, dict):
        for attr, branches in tree.items():
            for val, subtree in branches.items():
                print(f"{indent}{attr} = {val} ->")
                print_tree(subtree, indent + "   ")
    else:
        print(f"{indent}[{tree}]")

# Example dataset (Play Tennis dataset)
dataset = [
    {"Outlook": "Sunny", "Temp": "Hot", "Humidity": "High", "Wind": "Weak", "Play": "No"},
    {"Outlook": "Sunny", "Temp": "Hot", "Humidity": "High", "Wind": "Strong", "Play": "No"},
    {"Outlook": "Overcast", "Temp": "Hot", "Humidity": "High", "Wind": "Weak", "Play": "Yes"},
    {"Outlook": "Rain", "Temp": "Mild", "Humidity": "High", "Wind": "Weak", "Play": "Yes"},
    {"Outlook": "Rain", "Temp": "Cool", "Humidity": "Normal", "Wind": "Weak", "Play": "Yes"},
    {"Outlook": "Rain", "Temp": "Cool", "Humidity": "Normal", "Wind": "Strong", "Play": "No"},
    {"Outlook": "Overcast", "Temp": "Cool", "Humidity": "Normal", "Wind": "Strong", "Play": "Yes"},
    {"Outlook": "Sunny", "Temp": "Mild", "Humidity": "High", "Wind": "Weak", "Play": "No"},
    {"Outlook": "Sunny", "Temp": "Cool", "Humidity": "Normal", "Wind": "Weak", "Play": "Yes"},
    {"Outlook": "Rain", "Temp": "Mild", "Humidity": "Normal", "Wind": "Weak", "Play": "Yes"},
]

attributes = ["Outlook", "Temp", "Humidity", "Wind"]

# Build decision tree
tree = id3(dataset, attributes, "Play")

print(" Decision Tree (Text Visualization):\n")
print_tree(tree)
```

```python
# Sample data points
data = [
    (1, 2),
    (2, 3),
    (4, 5),
    (3, 4),
    (5, 6)
]

# Function to calculate the mean of a list
def mean(values):
    return sum(values) / len(values)

# Function to calculate the coefficients m (slope) and b (intercept)
def linear_regression(data):
    x_values = [point[0] for point in data]
    y_values = [point[1] for point in data]

    x_mean = mean(x_values)
    y_mean = mean(y_values)
    # Calculate numerator and denominator for slope (m)
    numerator = 0
    denominator = 0
    for x, y in data:
        numerator += (x - x_mean) * (y - y_mean)
        denominator += (x - x_mean) ** 2
    m = numerator / denominator
    b = y_mean - m * x_mean
    return m, b
# Function to predict y for a given x using y = mx + b
def predict(x, m, b):
    return m * x + b

# Running the linear regression
m, b = linear_regression(data)
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")

# Predicting new values
test_x = 6
predicted_y = predict(test_x, m, b)
print(f"Predicted y for x = {test_x} is {predicted_y}")
```

```python


import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules

# Example dataset (transactions)
dataset = [
    ['milk', 'bread', 'eggs'],
    ['milk', 'bread'],
    ['milk', 'eggs'],
    ['bread', 'butter'],
    ['milk', 'bread', 'butter', 'eggs']
]

# Convert dataset to one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

print("Transaction DataFrame:")
print(df)

# Apply FP-Growth algorithm
frequent_itemsets = fpgrowth(df, min_support=0.4, use_colnames=True)
print("\nFrequent Itemsets:")
print(frequent_itemsets)

# Generate Association Rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Convert frozensets to readable strings
rules['antecedents'] = rules['antecedents'].apply(lambda x: ', '.join(list(x)))
rules['consequents'] = rules['consequents'].apply(lambda x: ', '.join(list(x)))

print("\nAssociation Rules:")
print(rules[['antecedents','consequents','support','confidence','lift']])
```
